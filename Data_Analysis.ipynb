```python
#Imputation Function
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age


full_data['Age']=full_data[['Age','Pclass']].apply(impute_age, axis=1)
```

```python
full_data.drop('Cabin',axis=1,inplace=True)
```

```python
full_data.dropna(inplace = True)
```

```python
#remoive unneccessory columns
full_data.drop(['Name' , 'Ticket'], axis=1, inplace=True)
```

```python
#convert objects to category data type
objcat = ['Sex', 'Embarked']
for colname in objcat:
   full_data[colname] = full_data[colname].astype('category')
```

```python
full_data
```

```python
#remove Passengerid
full_data.drop('PassengerId', axis=1, inplace=True)
```

```python
#identify categorical features
full_data.select_dtypes(include=['category']).columns
```

```python
#convert categorical variables into Dummys or indicator variables
sex=pd.get_dummies(full_data['Sex'],drop_first=True)
embark=pd.get_dummies(full_data['Embarked'],drop_first=True)
```

```python
full_data.head(5)
```

```python
#Add Dummys column to data frame
full_data=pd.concat([full_data,sex,embark],axis=1)
full_data.head(5)
```

```python
X=full_data.drop(['Survived'],axis=1)
y=full_data['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```
